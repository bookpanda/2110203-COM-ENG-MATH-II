```
import numpy as np
import pandas as pd
from scipy import signal,fftpack
import cv2
from skimage.io import imread
import matplotlib.pyplot as plt
import IPython.display as ipd
import os

import librosa
import parselmouth
```

```
sampling_rate = 32000
N=10001
Nf = 3 # Nf--> num freq
t= np.arange(N,dtype=float)
# pick rand period betwwen 10-2010 and convert to freq

# random period
Ts = np.random.rand(Nf)*2000+10
fs=1/Ts

# fs in sampling rate = 32000
fs_real = fs*sampling_rate

# pick rand Amp and phase
amp = np.random.rand(Nf)*200+ 100
phi = np.random.rand(Nf)*2*np.pi

# create clean signal
h = np.zeros(N)
for i in range(len(fs)):
    h += amp[i]*np.sin(2*np.pi*fs[i]*t + phi[i])

# signal with noise
h_w_noise = h + np.random.randn(N)*3*h + np.random.randn(N)*700
```

```
# TODO 1.1 : plot (1) clean signal and (2) noisy signal with label
plt.figure(figsize=(14, 6))

# plot clean signal
plt.subplot(2, 1, 1)
plt.plot(t, h, label='Clean Signal')
plt.title('Clean Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)

# plot noisy signal
plt.subplot(2, 1, 2)
plt.plot(t, h_w_noise, label='Noisy Signal', color='r')
plt.title('Noisy Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```
# TODO 1.2: plot magnitude of FFT of the noisy signal (freq sort form min--> max)
fft_h_w_noise = np.fft.fft(h_w_noise)

frequencies = np.fft.fftfreq(N, 1/sampling_rate)
magnitude = np.abs(fft_h_w_noise)

sorted_indices = np.argsort(frequencies)
frequencies_sorted = frequencies[sorted_indices]
magnitude_sorted = magnitude[sorted_indices]

plt.figure(figsize=(10, 6))
plt.plot(frequencies_sorted, magnitude_sorted)
plt.title('Magnitude of FFT of Noisy Signal (Sorted by Frequency)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()

```

```
# TODO 1.3 : cleaning the noisy signal using magnitude of FFT
plt.figure(figsize=(14, 6))

# noisy signal
plt.subplot(2, 1, 1)
plt.plot(t, h_w_noise, label='Noisy Signal', color='r')
plt.title('Noisy Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()

for i in range(1, 10):
  ratio = 0.1*i
  threshold = np.max(magnitude) * ratio
  fft_h_w_noise_cleaned = np.where(magnitude > threshold, fft_h_w_noise, 0)
  h_cleaned = np.fft.ifft(fft_h_w_noise_cleaned).real

  # filtered signal
  plt.figure(figsize=(14, 6))
  plt.subplot(2, 1, 2)
  plt.plot(t, h_cleaned, label=f'Filtered Signal (threshold = {ratio} * max magnitude)', color='g')
  plt.title(f'Filtered Signal (After FFT Filtering, threshold = {ratio} * max magnitude)')
  plt.xlabel('Time')
  plt.ylabel('Amplitude')
  plt.grid(True)

  plt.tight_layout()
  plt.show()
```

```
# TODO 1.4 : plot clean signal, noise signal and filtered signal (from your result in TODO 3.3) with label
plt.figure(figsize=(14, 8))

# plot clean (original) signal
plt.subplot(3, 1, 1)
plt.plot(t, h, label='Clean Signal', color='b')
plt.title('Clean Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)

# plot noisy signal
plt.subplot(3, 1, 2)
plt.plot(t, h_w_noise, label='Noisy Signal', color='r')
plt.title('Noisy Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)

ratio = 0.4
threshold = np.max(magnitude) * ratio
fft_h_w_noise_cleaned = np.where(magnitude > threshold, fft_h_w_noise, 0)
h_cleaned = np.fft.ifft(fft_h_w_noise_cleaned).real

# plot filtered signal
plt.subplot(3, 1, 3)
plt.plot(t, h_cleaned, label='Filtered Signal', color='g')
plt.title(f'Filtered Signal (After FFT Filtering, threshold = {ratio} * max magnitude)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```
# TODO 1.5 : export with IPython.display, listen to (1) original signal (2) signal with noise (3) signal after filtered

# original clean signal
print("Original Clean Signal:")
ipd.display(ipd.Audio(h, rate=sampling_rate))

# noisy signal
print("Noisy Signal:")
ipd.display(ipd.Audio(h_w_noise, rate=sampling_rate))

# filtered signal
print("Filtered Signal (After FFT Cleaning):")
ipd.display(ipd.Audio(h_cleaned, rate=sampling_rate))
```

```
# TODO 1.6 : Write to explain and analyze the results
# The original signal was a single clear beat
# The noisy signal sounded like TVs without signal, it is unthinkable that the original is the sound of a single beat.
# The filtered signal sounded very similar to the orignal signal, thanks to FFT cleaning with chosen magnitude threshold of 0.4*max_threshold.
# Had the threshold been too low, the filtered sound would still have had some noise.
# Had the threshold been too high, the filtered sound would have been too different from the original noise (see the waves in 1.3).
```

Download a 1000 x 1000 image ("hamtaro.png") below

<a href="https://ibb.co/kHTjZbb"><img src="https://i.ibb.co/kHTjZbb/hamtaro.png" alt="hamtaro" border="0"></a>

```
screen_shot = cv2.imread('hamtaro.png',0)

plt.figure(figsize=(10,10))
plt.imshow(screen_shot, cmap='gray')
plt.show()
```

```
# Apply FFT to the given image
F1 = fftpack.fft2((screen_shot).astype(float))
F2 = fftpack.fftshift(F1) # FFT center zeros freq
plt.figure(figsize=(10,10))
plt.imshow( (20*np.log10( 0.1 + np.abs(F2))).astype(int), cmap=plt.cm.gray)
plt.show()
```

```
# TODO 2.1 : Implement an ideal high-pass filter with a box size of 100x100 pixels on the given image
n = 100

rows, cols = screen_shot.shape
crow, ccol = rows // 2, cols // 2  # image center

# mask with high-pass filter
mask = np.ones((rows, cols), np.float32)
mask[crow - n // 2:crow + n // 2, ccol - n // 2:ccol + n // 2] = 0

# apply mask to FFT of image
F2_filtered = F2 * mask

# invert 2D FFT of filtered image
F1_filtered = fftpack.ifftshift(F2_filtered)
img_filtered = np.abs(fftpack.ifft2(F1_filtered))

plt.figure(figsize=(10, 10))
plt.imshow(img_filtered, cmap='gray')
plt.title('Filtered image with ideal high-pass filter')
plt.show()
```

```
# TODO 2.2 : Implement an ideal low-pass filter with a box size of 100x100 pixels on the given image

# mask with low-pass filter
mask = np.zeros((rows, cols), np.float32)
mask[crow - n // 2:crow + n // 2, ccol - n // 2:ccol + n // 2] = 1

# apply mask to FFT of image
F2_filtered = F2 * mask

# invert 2D FFT of filtered image
F1_filtered = fftpack.ifftshift(F2_filtered)
img_filtered = np.abs(fftpack.ifft2(F1_filtered))

plt.figure(figsize=(10, 10))
plt.imshow(img_filtered, cmap='gray')
plt.title('Filtered image with ideal low-pass filter')
plt.show()
```
